```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
title = []
for i in range(40):
    title.append(str(i))
```

```python
# train dataset
df_train = pd.read_csv('train.csv',names = title) 
# train target
df_label = pd.read_csv('trainLabels.csv',names = ['class'])
# test dataset
df_test = pd.read_csv('test.csv',names = title)
```

```python
# let column can show all
pd.set_option('display.max_columns', None)
```

```python
df_train.describe()
```

```python
df_label.head(5)
```

```python
# split to test and validation
X_train,X_val, y_train, y_val = train_test_split(df_train,df_label,test_size=0.2, random_state=42)
```

```python
# build a linear regression model
lgregr = linear_model.LogisticRegression(solver = 'lbfgs')

# put the train data for training
lgregr.fit(X_train, y_train)

# validate
y_pred = lgregr.predict(X_val)
```

```python
ans = y_val.to_numpy()
```

# Baseline (doesn't do any processing)

```python
accuracy = 0

for i in range(len(y_val)):
    if y_pred[i] == ans[i][0]:
        accuracy = accuracy + 1
        
print("The baseline is : ",accuracy / len(y_val)*100.0 ,'%')
```

```python
MSEs = cross_val_score(lgregr,X_train, y_train, cv = 10)
```

```python
print(MSEs)
```

```python
print(np.mean(MSEs))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score

#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier( n_estimators = 60, max_features = 'sqrt',criterion = 'entropy')      
clr_rf = clf_rf.fit(X_train,y_train)

ac = accuracy_score(y_val,clf_rf.predict(X_val))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_val,clf_rf.predict(X_val))
sns.heatmap(cm,annot=True,fmt="d")
```

# KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Trying to find the best no of neighbours.
cross_val_accuracy_per_neighbours = []
for i in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    cross_val_accuracy = np.mean(cross_val_score(knn,X_train, y_train, cv=10))
    cross_val_accuracy_per_neighbours.append([i, cross_val_accuracy])
```

```python
print(cross_val_accuracy_per_neighbours.sort(reverse = True))
```

```python
print(cross_val_accuracy_per_neighbours)
```
